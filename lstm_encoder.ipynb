```python
SEQ_LEN     = 24 
N_FEATURES  = 8 
Z_DIM       = 32
NUM_SAMPLES = 400

model = LSTMAutoencoder(SEQ_LEN, N_FEATURES, Z_DIM)
```

```python
data = torch.from_numpy(np.random.rand(SEQ_LEN, 1, N_FEATURES).astype(np.float32))
```

```python

data = np.arange(0, SEQ_LEN)
data = np.tile(data, (N_FEATURES))
data = np.reshape(data, (SEQ_LEN, 1, N_FEATURES))

scale  = np.reshape(np.repeat(np.random.rand(N_FEATURES), SEQ_LEN), (SEQ_LEN, 1, N_FEATURES))
data = scale * data
data = np.tile(data, (1,  NUM_SAMPLES, 1  ))

offset = np.reshape(np.repeat(np.random.rand(N_FEATURES, NUM_SAMPLES), SEQ_LEN), (SEQ_LEN, NUM_SAMPLES, N_FEATURES))


data = data + offset
data = np.sin(data)
print(data.shape)
data = torch.from_numpy(data.astype(np.float32))
```

```python
loss_function = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.1)

for epoch in range(300):
    for idx in range(0, NUM_SAMPLES -20, 20):

        input = data[:,idx:idx+20, :]
        model.zero_grad()
        output = model(input)
        loss = loss_function(output, input)
        loss.backward()
        optimizer.step()

        print("Epoc: {:04d}-{:04d} Loss: {}".format(epoch, idx, loss.item()))


```
